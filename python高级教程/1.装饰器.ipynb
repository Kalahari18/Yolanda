# 装饰器
用来装饰其他函数，为其他函数添加功能
类方法，类名字直接调用
原则 
    装饰器不能修改被装饰函数的原码
    装饰器不能修改被装饰函数的调用方式
任务
编写一个装饰器，要求使用该装饰器能够统计任意函数运行时间~
函数即变量，高阶函数，嵌套函数
## 函数即变量
函数可以被调用，也可以作为函数名调用
函数名跟变量名一样，只是一个变量的标识符，他指向函数定义对应的内存地址
在函数定义中去调用其他函数时，并不会立即调用函数
在执行一个调用了其他函数的函数时，如果在内存中还没有找到被调用函数的定义，则程序会报错
## 装饰器-高阶函数
1.接受函数名作为形参
2.返回值中包含函数名
高阶函数对装饰器的意义
接受函数名作为形参（不改变被装饰函数的代码的前提下增加功能）
返回值中包含函数名（不改变装饰器函数的调用方式）
## 装饰器-嵌套函数
通过def关键字定义在另一个函数中的函数叫做嵌套函数

## 被装饰函数带参数
通过提供不定量参数来自适应被装饰函数
## 装饰器本身带参数
两层嵌套函数
## 被装饰函数带返回值
最里层函数返回

## 装饰函数与闭包
闭包的作用
可以用来在一个函数与一组私有变量之间创建关联关系，在给定函数被多次调用的过程中，这些私有变量能够保持其持久性（保存运行环境与变量的状态）

特征：
必须要有函数的嵌套，外层函数必须返回内层函数，外层啊哈双女户必须给包装起来的运行环境，在这个包装的运行环境里面，必须开始包装

# 多线程的高级应用
## 多线程及多进程编程
进程：相当于公司的老板，他自己本身不干活，但为干活的员工提供资源，平台
线程：相当于公司的员工，真正干活的人就是这些员工
线程和进程之间的关系
（1）一个线程只能属于一个进程，一个进程可以有多个线程，每个进程至少一个线程
（2）资源分配给进程，一个进程中的所有线程将共享该进程中所有资源
（3）线程在执行过程中，需要协作同步，不同进程间的线程需要利用同步机制来实现通讯
（4）真正执行任务的是线程，它是进程内可调度的实体
一个进程必须包含至少一个线程（火锅桌子至少一个人）
一个进程可以包含多个线程（桌子多个人）
不同进程间数据很难共享（共享同一桌子上的菜）
同一进程中，不同线程间的数据则很同一共享（同一桌子上的菜很容易共享）
创建进程要比创建线程消耗更多的计算机资源（多加一个桌子比多加一个人成本开销大很多）
进程间不会相互影响，而相同进程内部如果一个线程出现异常，则其他所有线程则全部挂掉
同一进程内不同线程使用同一个资源时需要加锁

多进程的优点：

多线程优点：效率比较高，不会耗费大量资源

多进程使用场景：对于计算密集型任务，比较适合多进程
多线程使用场景：适合io密集型任务，比如文件读取以及io操作等
